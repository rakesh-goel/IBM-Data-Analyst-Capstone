## Distribution


### Determine how the data is distributed


The column `ConvertedComp` contains Salary converted to annual USD salaries using the exchange rate on 2019-02-01.

This assumes 12 working months and 50 working weeks.


Plot the distribution curve for the column `ConvertedComp`.


```python
# your code goes here
#NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays
import numpy as np
# Matplotlib is a plotting library for python and pyplot gives us a MatLab like plotting framework. We will use this in our plotter function to plot data.
import matplotlib.pyplot as plt
#Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics
import seaborn as sns

df['ConvertedComp'].describe()

```

Dropping all the rows having NaNs in the column ConvertedComp of the dataframe 

```python
df_convertedcomp = df.dropna(subset=["ConvertedComp"], axis=0, inplace=False)
```

Plotting the distribution curve

```python
sns.displot(df_convertedcomp['ConvertedComp'], kind= "kde", color="r", aspect=2)

plt.ticklabel_format(style='plain', axis='x',useOffset=False)

plt.title('Distribution of Annual Salary')
plt.xlabel('Annual Salary in USD')
plt.ylabel('Proportion of Respondents')

plt.show()
```

Plot the histogram for the column `ConvertedComp`.


```python
# your code goes here
sns.displot(df_convertedcomp['ConvertedComp'], kind= "hist", color="r", aspect=2)

plt.ticklabel_format(style='plain', axis='x',useOffset=False)

plt.title('Histogram of Annual Salary')
plt.xlabel('Annual Salary in USD')
plt.ylabel('Number of Respondents')

plt.show()
```

## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here

df_convertedcomp['ConvertedComp'].describe()
```

```python
df_convertedcomp['ConvertedComp'].plot(kind='box', figsize=(8, 10))

plt.ticklabel_format(style='plain', axis='y',useOffset=False)
plt.title('Box plot of converted compensation(Salary)')
plt.ylabel('Salary')

plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
Q3 = df_convertedcomp['ConvertedComp'].quantile(0.75)
Q1 = df_convertedcomp['ConvertedComp'].quantile(0.25)
print(Q3, Q1)
IQR = Q3 - Q1
print(IQR)
```

Find out the upper and lower bounds.


```python
# your code goes here
l_bound = Q1-(IQR*1.5)
u_bound = Q3+(IQR*1.5)

print(u_bound,l_bound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here

df_convertedcomp[(df_convertedcomp['ConvertedComp'] > u_bound) | (df_convertedcomp['ConvertedComp'] < l_bound)].shape[0]
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_clean = df_convertedcomp[df_convertedcomp['ConvertedComp'] <= u_bound]
# df_new.drop(['level_0', 'index'], axis=1, inplace=True)
df_clean['ConvertedComp'].describe()
```

```python
df_clean['ConvertedComp'].plot(kind='box', figsize=(8, 10))

plt.ticklabel_format(style='plain', axis='y',useOffset=False)
plt.title('Box plot of converted compensation(Salary) - outbound removed')
plt.ylabel('Salary')

plt.show()
```
